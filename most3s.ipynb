### Loop through all the tables and rows

```python
tables = soup.find_all("table")
len(tables)
```

```python
for table in tables:
    print(len(table.find_all("tr")))
```

```python
data = list()
for table in tables:
    for row in table.find_all("tr")[1:]:
        info_str = row.get_text(",", strip=True)
        info_arr = info_str.split(",")
        if info_arr[2] == "(Pl)":
            playoffs = True
            info = info_arr[:2] + info_arr[3:5] + [playoffs]
        else:
            playoffs = False
            info = info_arr[:4] + [playoffs]
        data.append(info)

data[:10], len(data)
```

### Create pandas DataFrame

```python
import pandas as pd

columns = ["Made3s", "Player", "Team", "Season", "Playoffs"]

most_in_game = pd.DataFrame(data=data, columns=columns)
most_in_game.head()
```

```python
most_in_game.info()
```
